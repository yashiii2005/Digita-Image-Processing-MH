```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read two images (make sure same size)
img1 = cv2.imread(r'C:\Users\LENOVO\OneDrive\Desktop\miranda\a.png')
img2 = cv2.imread(r'C:\Users\LENOVO\OneDrive\Desktop\miranda\b.png')
```

```python
img1 = cv2.resize(img1, (300, 300))
#img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

img2 = cv2.resize(img2, (300, 300))
#img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
# Display images
plt.subplot(1,2,1)
plt.title('Image 1')
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Image 2')
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
#Addition of two images

# Manual arithmetic addition
added_manual = img1.astype(np.int16) + img2.astype(np.int16)
added_manual = np.clip(added_manual, 0, 255).astype(np.uint8)

# Using OpenCV built-in
added_cv = cv2.add(img1, img2)

# Show both
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Manual Addition")
plt.imshow(added_manual, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("cv2.add()")
plt.imshow(added_cv, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Manual subtraction
sub_manual = img1.astype(np.int16) - img2.astype(np.int16)
sub_manual = np.clip(sub_manual, 0, 255).astype(np.uint8)

# Using OpenCV
sub_cv = cv2.subtract(img1, img2)

# Show both
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Manual Subtraction")
plt.imshow(sub_manual, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("cv2.subtract()")
plt.imshow(sub_cv, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Manual mean
mean_manual = np.sum(img1) / (img1.shape[0] * img1.shape[1])

# OpenCV mean (returns tuple for each channel)
mean_cv = cv2.mean(img1)[0]

print("Manual Mean Value:", mean_manual)
print("OpenCV Mean Value:", mean_cv)
```

```python
# Increase brightness manually
bright_manual = img1 + 50
bright_manual = np.clip(bright_manual, 0, 255).astype(np.uint8)

# Decrease brightness manually
dark_manual = img1 - 50
dark_manual = np.clip(dark_manual, 0, 255).astype(np.uint8)

# Using OpenCV add/subtract
bright_cv = cv2.add(img1, 50*np.ones(img1.shape, dtype=np.uint8))
dark_cv = cv2.subtract(img1, 50*np.ones(img1.shape, dtype=np.uint8))

# Display all
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Brighter (+50)")
plt.imshow(bright_cv, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Darker (-50)")
plt.imshow(dark_cv, cmap='gray')
plt.axis('off')
plt.show()
```
