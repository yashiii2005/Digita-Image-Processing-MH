```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read image
image = cv2.imread(r'/home/pc-cs/Desktop/yashasvi1315/DIP/sml.png')
plt.imshow(image)
```

```python
image.shape
```

```python
# Resize the image to 300x300 pixels
image = cv2.resize(image, (100, 100))
```

```python
# Convert the color image to grayscale
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
image.shape
```

```python
# Manual negative transformation: new_pixel = 255 - original_pixel
negative_image = 255 - image
plt.imshow(negative_image)
```

```python
# Manual horizontal flip by reversing columns
flip_image = image[:, ::-1]
plt.imshow(flip_image)
```

```python
# c. Thresholding (convert to grayscale first)
threshold_value = 127
threshold_image = np.zeros_like(image, dtype=np.uint8)  # create empty output image

# For each pixel, if pixel value >= threshold_value, set to 255 else 0
threshold_image[image >= threshold_value] = 255
plt.imshow(threshold_image)
```

```python
# d. Contrast stretching on grayscale image
in_min = np.min(image)
in_max = np.max(image)
out_min = 0
out_max = 255

contrast_stretched_image = ((image - in_min) * ((out_max - out_min) / (in_max - in_min)) + out_min).astype(np.uint8)

plt.imshow(contrast_stretched_image)
```

```python
# Display the individual channels
plt.subplot(1, 5, 1)
plt.imshow(image, cmap='gray')
plt.title("original")
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(negative_image)
plt.title("negative")
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(flip_image)
plt.title("flip")
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(threshold_image)
plt.title("threshold")
plt.axis('off')

plt.subplot(1, 5, 5)
plt.imshow(contrast_stretched_image)
plt.title("contrast")
plt.axis('off')
```
