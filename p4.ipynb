```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread(r'C:\Users\LENOVO\OneDrive\Desktop\miranda\a.png')
img2 = cv2.imread(r'C:\Users\LENOVO\OneDrive\Desktop\miranda\b.png')
```

```python
img1 = cv2.resize(img1, (300, 300))
img2 = cv2.resize(img2, (300, 300))
```

```python
# Display images
plt.subplot(1,2,1)
plt.title('Image 1')
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Image 2')
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Manual AND using numpy
and_manual = np.bitwise_and(img1, img2)

# OR using OpenCV function
and_cv = cv2.bitwise_and(img1, img2)

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Manual AND")
plt.imshow(and_manual, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("cv2.bitwise_and()")
plt.imshow(and_cv, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Manual OR
or_manual = np.bitwise_or(img1, img2)

# Using OpenCV
or_cv = cv2.bitwise_or(img1, img2)

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Manual OR")
plt.imshow(or_manual, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("cv2.bitwise_or()")
plt.imshow(or_cv, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Intersection is same as AND
intersection = cv2.bitwise_and(img1, img2)

plt.imshow(intersection, cmap='gray')
plt.title("Intersection of Two Images")
plt.axis('off')
plt.show()
```

```python
watermarked = cv2.bitwise_xor(img1, img2)

# Display images
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Original Image 1")
plt.imshow(img1, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Watermark (Image 2)")
plt.imshow(img2, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Watermarked Image")
plt.imshow(watermarked, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Manual NOT (negative)
not_manual = 255 - img1

# Using OpenCV
not_cv = cv2.bitwise_not(img1)

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Manual NOT (Negative)")
plt.imshow(not_manual, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("cv2.bitwise_not()")
plt.imshow(not_cv, cmap='gray')
plt.axis('off')
plt.show()
```
