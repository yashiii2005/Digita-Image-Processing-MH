```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#image in grayscale
img = cv2.imread(r'C:\Users\LENOVO\OneDrive\Desktop\miranda\b.png', cv2.IMREAD_GRAYSCALE)
```

Histogram Calculation & Equalization Using Inbuilt Function

```python
# Histogram Calculation
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Histogram Equalization
equalized = cv2.equalizeHist(img)
```

```python
#Display results
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.plot(hist, color='black')
plt.title('Original Histogram')

plt.subplot(2,2,3)
plt.imshow(equalized, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.subplot(2,2,4)
plt.hist(equalized.ravel(), 256, [0,256], color='black')
plt.title('Equalized Histogram')

plt.tight_layout()
plt.show()
```

Histogram Calculation & Equalization Without Inbuilt Function

```python
# Calculate Histogram manually
hist = np.zeros(256)
for pixel in img.ravel():
    hist[pixel] += 1

# Normalize Histogram
hist_norm = hist / hist.sum()

#Compute Cumulative Distribution Function (CDF) 
cdf = hist_norm.cumsum()

#Equalization formula
equalized = np.floor(255 * cdf[img]).astype(np.uint8)

# Calculate Equalized Histogram
hist_eq = np.zeros(256)
for pixel in equalized.ravel():
    hist_eq[pixel] += 1
```

```python
# Display
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.plot(hist, color='black')
plt.title('Original Histogram')

plt.subplot(2,2,3)
plt.imshow(equalized, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.subplot(2,2,4)
plt.plot(hist_eq, color='black')
plt.title('Equalized Histogram')

plt.tight_layout()
plt.show()
```
