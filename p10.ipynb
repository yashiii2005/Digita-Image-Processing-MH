```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load image in grayscale
img = cv2.imread(r'C:\Users\LENOVO\OneDrive\Desktop\miranda\img.png', 0)   # keep an image named "image.jpg" in the same folder

# Check if image loaded correctly
if img is None:
    print("Error: Image not found. Please check the file name or path.")
else:
    plt.imshow(img, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")
```

Sobel Edge Detection

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobelx, sobely)

plt.imshow(sobel_combined, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis("off")
```

Prewitt Edge Detection

```python
prewittx = np.array([[-1,0,1],
                     [-1,0,1],
                     [-1,0,1]])

prewitty = np.array([[-1,-1,-1],
                     [0,0,0],
                     [1,1,1]])

prewitt_x = cv2.filter2D(img, -1, prewittx)
prewitt_y = cv2.filter2D(img, -1, prewitty)
prewitt_combined = prewitt_x + prewitt_y

plt.imshow(prewitt_combined, cmap='gray')
plt.title("Prewitt Edge Detection")
plt.axis("off")
```

Roberts Edge Detection

```python
robertsx = np.array([[1, 0],
                     [0, -1]])

robertsy = np.array([[0, 1],
                     [-1, 0]])

roberts_x = cv2.filter2D(img, -1, robertsx)
roberts_y = cv2.filter2D(img, -1, robertsy)
roberts_combined = roberts_x + roberts_y

plt.imshow(roberts_combined, cmap='gray')
plt.title("Roberts Edge Detection")
plt.axis("off")
```

Laplacian Edge Detection

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)

plt.imshow(laplacian, cmap='gray')
plt.title("Laplacian")
plt.axis("off")
```

Canny Edge Detection

```python
canny = cv2.Canny(img, 50, 150)

plt.imshow(canny, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
```
